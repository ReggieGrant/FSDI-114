OOP - Object Oriented Programming 

Python is an object-oriented language, allowing us to structure code using classes and objects for better organization and reusability.

### Advantages:
- Provides a clear structure to programs
- Makes code easier to maintain, reuse and debug
- Helps keep your code DRY (Do not Repeat Yourself)
- Allows to build reusable applictions with less code


### Classes and Objects
A class defines what an object should look like and it's behavior.

An object is created bases on that class.

For example:

Class -> Fruit

Objects -> Apple, Banana, Mango

Class -> Audi, Tesla, Toyota

### The four pillars of OOP
1. Encapsulation
2. Abstraction
3. Inheritance
4. Polymorphism

```python
# Syntax
# Rerved keyword 'class' + name of the class 
# Note: the name of the class should follow the Pascal Case notation


# Constructor - Is always executed when the class is being initiated
# Properties/Attributes - the variables that helps us describe our class
class MyClass:
    x = 5 #This is a property
    y = 10

p1 = MyClass() #Ox000A
p2 = MyClass() #Ox000B
print(p1.x)
print(p1.y)
print(p2.x)

```

```python
class Person:
    #Constructor - Executes when initializing and object
    def __init__(self,name,last_name,age):
        self.name = name
        self.last_name = last_name
        self.age = age

    def greeting(self):
        return f"Hello from {self.name}!"

    def __str__(self):
        return f"Person (name={self.name}, last_name={self.last_name}, age={self.age})"

person1 = Person("Reggie","Grant",38)
person2 = Person("Name","NAME", 99)

print (person1.name)
print (person2.name)
print (person1.greeting())
print (person2.greeting())
print (person2)
```

```python
class Calculator:
    def __init__(self, x,y):
        self.x = x
        self.y = y

    def add(self):
        return self.x + self.y

    def multiply(self):
        return self.x * self.y

    def subtract(self):
        return self.x - self.y

    def divide(self):
        return self.x % self.y

    def __str__(self):
        return f"Calculator Obj (x={self.x},y={self.y})"



                

c1 = Calculator(10,2)
c1.add()
c1.subtract()
c1.divide()
c1.multiply
```
