```python
# Singly Linked-Lists

# Node -> [data | link]
"""
0x00500            0x00A1            0x00C5            0x00D7            0x00C0
[298 | 0x00a1] -> [305 | 0x00C5] -> [320 | 0x00D7] -> [301 | 0x00C0] -> [292 | null]

Inserting element 284 at position 1

0x00500            0x0002             0x00A1            0x00CS            0x00D7            0x00C0
[298 | 0x00A1] -> [284 | 0x00A1] -> [305 | 0x00CS] -> [320 | 0x00D7] -> [301 | 0x00C0] -> [292 | null]

Insert Element at beginning = O(1)
Delete Element at beginning = O(1)
Insert/Delete Element at the end = O(n)
Linked List Traversal = O(n)
Accessing Element by value = O(n)

"""
class Node:
    def __init__(self, data, link):
        self.data = data
        self.link = link

    def __str__(self):
        return f"Node ( data = {self.data}, link = {self.link})"

n1 = Node(298, None)
print(n1)
```
