## Find the total of numbers from a collection of numbers that divide by 2 the module is 0

```
Examples:
1. Given [3,4,10,15,999,2] Then the output should be: total = 3
2. Given [2,4,6,10,3.7] Then the output should be: total = 4
```
### Criteria
1. Create a function that receives a collection and return a interger with the total.

### Hint
- '%' is the arithmetic operand to get the module from a division
- 2 % 2 = 0


The collections of numbers is: **numbers = [3,5,27.20.6.3.5,60]**

````python
numbers = [3,5,27,20,6,3.5,60]

````

*Pseudocode* is like coding without worrying about the syntax and the rules from a programming language

START.

create a function called my_function.

    initialize a variable to store the amount of even numbers.
    
    travel the array/list so I can watch each value.
    
        make a comparison if the actual value hads a module of 0.
        
            if this is right I would add one more to my variable.

````python
numbers = [3,5,27,20,6,3.5,60]

def my_function(x):
    total = 0
    for number in x:
        if number % 2 == 0:
            total = total + 1

    return total

my_function(numbers)    
````

# Problem 2

## Given 2 arrays, create a function that let the user know (true/false) wether these two arrays contain any common items.

```
Example:
array1 = ['a','b','c','x']
array2 = ['z','y','i']
should return False

array1 = ['a','b','c','x']
array2 = ['z','y','x']
should return True

````python
array1 = ['a','b','c','x'] #O(1)
array2 = ['z','y','x'] #O(1)

def comparing_arrays(arr1, arr2): #O(1)
    has_something_in_common = False #O(1)
    for item in arr1: #O(n)
        for item2 in arr2: #O(n) -> O(n^2)
            if item2 == item: #O(1)
                has_something_in_common = True #O(1)

    return has_something_in_common #O(1)

comparing_arrays(array1, array2) #O(1)
# This algorithm is O(n^2)
````

array1 = ['a','b','c','x'] #O(1)
array2 = ['z','y','x'] #O(1)

START
Loop through the first array
    create an object(dictionary)
        properties are the same as the items
        
loop through the second array and check if the items exsts in the creayed object
END


````python
array1 = ['a','b','c','x'] #O(1)
array2 = ['z','y','x'] #O(1)

def comparing_arrays_2(arr1, arr2): #O(1)
    # Loop throught first array and create a dictionary where properties == items in the array
    dict1 = {} #O(1)
    for item in arr1: #O(n)
        if dict1.get(item) == None:
            dict1[item] = True

    # Loop through the second array and check if item in the second one exists on created dictionary
    for item2 in arr2: #O(n)
        if dict1.get(item2) != None:
            return True

        return False

comparing_arrays_2(array1, array2)
# This algorithm has a complexity of O(n)
````
