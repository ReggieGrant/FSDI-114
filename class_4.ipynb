## Stacks 
A Stack is an ordered collection of elements where items are added and removed from one end, comonly referred to as the "top". The ordering principle id referred to as LIFO (last in, first out).

## Queues 
A queue ia an ordered collection of elements where items are added and removed from **opposite** ends, commonly referred to as the "front" and "back" or "rear" of the queue. The ordering principle is referred to as FIFO (first in, first out) of "first come first served".

````python


def invert_string(word):
    word_length = len(word)
    inverted = ""
    for index in range(1, len(word)+1):
        inverted += word[len(word)-index]
    return inverted 

invert_string("house")    
````

````python
class Stack:
    def __init__(self):
        self.items = []

    def push(self,value):
        self.items.append(value)

    def pop(self):
        return self.items.pop()

    def size(self):
        return len(self.items)

    def peek(self):
        return self.items[len(self.items)-1]

    def is_empty(self):
        return self.items == []

    def __str__(self):
        return f"Stack({self.items})"
````

````python
st1 = Stack()
st1.push(10)
st1.push(11)
st1.push(23)
st1.push(50)
print(st1)

print(st1.peek())
````

````python
# Re do the problem 3 and use Stacks

def inverting_strings_using_stacks(word):
    inverted = ""
    st1 = Stack()

    for letter in word:
        st1.push(letter)

    while not st1.is_empty():
        inverted += st1.pop()
        

    return inverted

inverting_strings_using_stacks("house")
````

````python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, element):
        self.items.insert(0,element)

    def dequeue():
        return self.items.pop()

    def peek(self):
        return self.items[len(self.items)-1]

    def size(self):
        return len(self.items)

    def is_empty(self):
        return self.items == []

    def __str__(self):
        return f"Queue({self.items})"

q1 = Queue()
q1.enqueue(10)
q1.enqueue(20)
q1.dequeue()
print(q1)
````
